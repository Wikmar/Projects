# Visualising Earnings Based On College Majors
We explore a cleaned dataset of outcomes of students who graduated from college between 2010 and 2012. We attempt to answer some questions such as:
1. Do popular major students make more money?
2. How many majors are predominantly male/female?
3. Which category of majors have the most students?

We start by setting up the environment and removing any rows containing null values.

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
recent_grads = pd.read_csv('datasets/recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape
cleaned_data_count
```

# Exploring Relations by Generating Scatter Plots

```python
recent_grads.plot('Sample_size', 'Median', kind='scatter')
```

```python
recent_grads.plot('Sample_size', 'Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot('Full_time', 'Median', kind='scatter')
```

We can deduce there is a clear cluster of jobs lying in a certain range with similar median values. As we branch out from these clusters the number of data points grows smaller. This indicates that certain jobs have higher median values. Sample size increasing does not have much effect on our data, apart from ensuring that the cluster of data we're observing is accurate even when sample size is high.

```python
recent_grads.plot('ShareWomen', 'Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot('Men', 'Median', kind='scatter')
```

```python
recent_grads.plot('Women', 'Median', kind='scatter')
```

We observe less data shown in women to median scatter plot, indicating more data lying in the cluster or less data available. We can't visibly determine if there is any difference between women and male differences.

# Bar Plots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

In the top 10 Majors, Astronomy and Astrophysics was the greatest share of women exceeding over half of all employed with that major. However when we look at the bottom 10 Majors, women far exceed over half of those employed with those Majors, with all 10 exceeding Astronomy and Astrophysics. 

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[-10:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

On average the bottom 10 Majors have a higher unemployment rate than the top 10, but there are of course some outliers that are visible in the bar plots. The reason this is is most likely that the more higher paying a Major is, businesses seek these skills learnt and are more difficult to attract potential employees, thus raise the wage so to attract the best minds to their business, causing a lower unemployment rate.
